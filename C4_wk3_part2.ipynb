Fetch the saved csv fro the neighborhoods from part1.

```python
nb2=pd.read_csv('neighborhoods_part1.csv')
nb2
```

```python
nb2['PostalCode'].to_csv('postal_codes.csv')
len(nb2)

Latitude = []
Longitude = []

for i in range(0,len(nb2)):
    lat_lng_coords = None
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(nb2.loc[i,'PostalCode']))
        lat_lng_coords = g.latlng
        %%time
    print(i,lat_lng_coords)
    Latitude.append(lat_lng_coords[0])
    Longitude.append(lat_lng_coords[1]) 
```

Save the coordinates, since the geocoder is very unstable.

```python
d2 = {'Latitude': Latitude, 'Longitude': Longitude}

coords = pd.DataFrame(d2)
coords

coords.to_csv('coords.csv')
```

Load the saved coordinates from file coords.csv (saved in previous step)

```python
saved_coords = pd.read_csv('coords.csv')
coords = saved_coords.loc[:,'Latitude':'Longitude']
coords
```

Insert the coordinates in the neighborhoods dataframe:

```python
nb3 = nb2

nb3['Latitude']=coords['Latitude']
nb3['Longitude']=coords['Longitude']
nb3 = nb3.drop(nb3.columns[0], axis=1)
nb3
```

Select only Boroughs containing the word 'Toronto' :

```python
nb_tor = nb3[(nb3['Borough'].str.contains('Toronto'))]
nb_tor
```

Save the final dataframe:

```python
nb_tor.to_csv('neighborhoods_coordinates_part2.csv')
```
